```python
app_events = pd.read_csv("app_events.csv")
complaints = pd.read_csv("complaints.csv")
customers = pd.read_csv("customers.csv")
data_dictionary = pd.read_csv("data_dictionary.csv")
deliveries = pd.read_csv("deliveries.csv")
drivers = pd.read_csv("drivers.csv")
hubs = pd.read_csv("hubs.csv")
incidents = pd.read_csv("incidents.csv")
orders = pd.read_csv("orders.csv")
vehicles = pd.read_csv("vehicles.csv")
```

```python
orders["order_created_at"] = pd.to_datetime(orders["order_created_at"])
deliveries["dispatch_time"] = pd.to_datetime(deliveries["dispatch_time"])
deliveries["delivery_completed_at"] = pd.to_datetime(deliveries["delivery_completed_at"])
complaints["created_at"] = pd.to_datetime(complaints["created_at"])
app_events["event_timestamp"] = pd.to_datetime(app_events["event_timestamp"])
incidents["reported_at"] = pd.to_datetime(incidents["reported_at"])
vehicles["commission_date"] = pd.to_datetime(vehicles["commission_date"])
customers["signup_date"] = pd.to_datetime(customers["signup_date"])
```

```python
orders_deliveries = pd.merge(
    orders,
    deliveries,
    on="order_id",
    how="left"
)

orders_deliveries.head()
```

```python
orders_deliveries_customers = pd.merge(
    orders_deliveries,
    customers,
    on="customer_id",
    how="left"
)

orders_deliveries_customers.head()
```

```python
full_service_data = pd.merge(
    orders_deliveries_customers,
    drivers,
    on="driver_id",
    how="left"
)

full_service_data.head()
```

```python
full_service_data = pd.merge(
    full_service_data,
    hubs,
    on="hub_id",
    how="left"
)

full_service_data.head()
```

```python
print("Orders rows:", len(orders))
print("Deliveries rows:", len(deliveries))
print("Combined orders-deliveries rows:", len(orders_deliveries))
print("Full service data rows:", len(full_service_data))
```

```python
print("Duplicate rows in full service data:", full_service_data.duplicated().sum())
print("Unique order IDs in orders:", orders["order_id"].nunique())
print("Unique order IDs in combined data:", full_service_data["order_id"].nunique())
```

```python
missing_values = full_service_data.isnull().sum().sort_values(ascending=False)
missing_values.head(20)
```

```python
missing_percent = (full_service_data.isnull().mean() * 100).sort_values(ascending=False)
missing_percent.head(20)
```

```python
clean_data = full_service_data.copy()

numeric_cols = clean_data.select_dtypes(include=["int64", "float64"]).columns
categorical_cols = clean_data.select_dtypes(include=["object"]).columns

for col in numeric_cols:
    clean_data[col] = clean_data[col].fillna(clean_data[col].median())

for col in categorical_cols:
    clean_data[col] = clean_data[col].fillna("Unknown")

clean_data.isnull().sum().sort_values(ascending=False).head(20)
```

```python
clean_data["is_failed_or_delayed"] = clean_data["delivery_status"].isin(["Failed", "Delayed"]).astype(int)

clean_data["delivery_duration_hours"] = (
    clean_data["delivery_completed_at"] - clean_data["dispatch_time"]
).dt.total_seconds() / 3600

clean_data["cost_per_km"] = clean_data["fuel_or_charge_cost"] / clean_data["route_distance_km"]

clean_data["high_value_order"] = np.where(clean_data["order_value"] >= clean_data["order_value"].quantile(0.75), 1, 0)

clean_data[[
    "order_id",
    "delivery_status",
    "is_failed_or_delayed",
    "delivery_duration_hours",
    "cost_per_km",
    "high_value_order"
]].head()
```

```python
wanted_numeric_cols = [
    "order_value",
    "promised_window_hours",
    "route_distance_km",
    "manual_route_override_count",
    "proof_of_completion_score",
    "customer_rating_post_delivery",
    "fuel_or_charge_cost",
    "delivery_duration_hours",
    "cost_per_km",
    "loyalty_score",
    "app_engagement_score",
    "training_score",
    "driver_rating",
    "capacity_score"
]

existing_numeric_cols = [col for col in wanted_numeric_cols if col in clean_data.columns]

print("Columns used for analysis:")
print(existing_numeric_cols)

print("\nColumns missing from dataset:")
print([col for col in wanted_numeric_cols if col not in clean_data.columns])

business_numeric = clean_data[existing_numeric_cols]

business_stats = business_numeric.describe().T
business_stats
```

```python
selected_columns = [
    "order_value",
    "route_distance_km",
    "fuel_or_charge_cost",
    "customer_rating_post_delivery",
    "manual_route_override_count"
]

for col in selected_columns:
    print("Column:", col)
    print("Mean:", np.mean(clean_data[col]))
    print("Median:", np.median(clean_data[col]))
    print("Standard Deviation:", np.std(clean_data[col]))
    print("Minimum:", np.min(clean_data[col]))
    print("Maximum:", np.max(clean_data[col]))
    print("25th Quantile:", np.quantile(clean_data[col], 0.25))
    print("75th Quantile:", np.quantile(clean_data[col], 0.75))
    print("-" * 50)
```

```python
zone_failure = clean_data.groupby("zone").agg(
    total_deliveries=("delivery_id", "count"),
    failed_or_delayed=("is_failed_or_delayed", "sum"),
    failure_delay_rate=("is_failed_or_delayed", "mean"),
    avg_rating=("customer_rating_post_delivery", "mean"),
    avg_cost=("fuel_or_charge_cost", "mean")
).reset_index()

zone_failure["failure_delay_rate_percent"] = zone_failure["failure_delay_rate"] * 100

zone_failure.sort_values(by="failure_delay_rate_percent", ascending=False)
```

```python
hub_performance = clean_data.groupby(["hub_id", "hub_name", "hub_type", "zone"]).agg(
    total_deliveries=("delivery_id", "count"),
    avg_capacity_score=("capacity_score", "mean"),
    avg_rating=("customer_rating_post_delivery", "mean"),
    avg_cost=("fuel_or_charge_cost", "mean"),
    failed_or_delayed=("is_failed_or_delayed", "sum")
).reset_index()

hub_performance["failure_delay_rate_percent"] = (
    hub_performance["failed_or_delayed"] / hub_performance["total_deliveries"]
) * 100

hub_performance.sort_values(by="failure_delay_rate_percent", ascending=False)
```

```python
driver_performance = clean_data.groupby(["driver_id", "base_zone", "employment_type"]).agg(
    total_deliveries=("delivery_id", "count"),
    avg_training_score=("training_score", "mean"),
    avg_driver_rating=("driver_rating", "mean"),
    avg_customer_rating=("customer_rating_post_delivery", "mean"),
    total_manual_overrides=("manual_route_override_count", "sum"),
    avg_manual_overrides=("manual_route_override_count", "mean"),
    failure_delay_rate=("is_failed_or_delayed", "mean")
).reset_index()

driver_performance["failure_delay_rate_percent"] = driver_performance["failure_delay_rate"] * 100

driver_performance.sort_values(by="total_manual_overrides", ascending=False).head(15)
```

```python
complaint_orders = pd.merge(
    complaints,
    orders,
    on=["order_id", "customer_id"],
    how="left"
)

complaint_analysis = complaint_orders.groupby(["service_type", "severity"]).agg(
    total_complaints=("complaint_id", "count"),
    avg_resolution_days=("resolution_days", "mean"),
    total_compensation=("compensation_amount", "sum"),
    avg_compensation=("compensation_amount", "mean")
).reset_index()

complaint_analysis.sort_values(by="total_compensation", ascending=False)
```

```python
app_failure = app_events.groupby(["event_type", "device_type", "zone_context"]).agg(
    total_events=("event_id", "count"),
    failed_events=("success_flag", lambda x: (x == 0).sum()),
    avg_api_latency=("api_latency_ms", "mean")
).reset_index()

app_failure["app_failure_rate_percent"] = (
    app_failure["failed_events"] / app_failure["total_events"]
) * 100

app_failure.sort_values(by=["app_failure_rate_percent", "avg_api_latency"], ascending=False).head(20)
```

```python
incident_summary = incidents.groupby(["incident_type", "severity", "resolution_status"]).agg(
    total_incidents=("incident_id", "count"),
    avg_resolved_hours=("resolved_hours", "mean"),
    max_resolved_hours=("resolved_hours", "max")
).reset_index()

incident_summary.sort_values(by="total_incidents", ascending=False)
```

```python
pairwise_cols = [
    "order_value",
    "promised_window_hours",
    "route_distance_km",
    "manual_route_override_count",
    "proof_of_completion_score",
    "customer_rating_post_delivery",
    "fuel_or_charge_cost",
    "delivery_duration_hours",
    "cost_per_km",
    "loyalty_score",
    "app_engagement_score",
    "training_score",
    "driver_rating",
    "capacity_score"
]

existing_pairwise_cols = [col for col in pairwise_cols if col in clean_data.columns]

pairwise_data = clean_data[existing_pairwise_cols].copy()

pairwise_data.head()
```

```python
app_zone_failure = app_events.groupby("zone_context").agg(
    total_events=("event_id", "count"),
    failed_events=("success_flag", lambda x: (x == 0).sum()),
    avg_latency=("api_latency_ms", "mean")
).reset_index()

app_zone_failure["failure_rate_percent"] = (
    app_zone_failure["failed_events"] / app_zone_failure["total_events"]
) * 100

plt.figure(figsize=(9, 5))
sns.barplot(
    data=app_zone_failure.sort_values(by="failure_rate_percent", ascending=False),
    x="failure_rate_percent",
    y="zone_context"
)
plt.title("App Failure Rate by Zone")
plt.xlabel("App Failure Rate (%)")
plt.ylabel("Zone")
plt.show()
```
